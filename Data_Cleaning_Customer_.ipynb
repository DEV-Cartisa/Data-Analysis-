```python
df[['Street_Address', 'State','Zip_Code' ]] =  df['Address'].str.split(',',2, expand = True )
df
# Split Street Address , into 3 columns
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
df
```

```python
df = df.replace('Na', '')
df = df.replace('N/a', '')
df
# NaN cannot be replaceed like that , as it is , Not A Number , its empty
df= df.fillna('')
df
```

```python
# Removing Columns , that dont want to be contacted
  # removing contact numbers , cause 'Do Not Contact' was ticked YES
  # we dont want to give - these numbeers to the - call center

# Loop - Drop column - with No
```

```python
for x in df.index:
  if df.loc[x, 'Do_Not_Contact'] == 'Y':
    df.drop(x, inplace=True)
df
```

```python
for x in df.index:
  if df.loc[x, 'Phone_Number'] == '':
    df.drop(x, inplace=True)
df


# if no phone number, remove
# other method
# df = df.drop.na((subset='Phone_Number'), inplace=True)
```
